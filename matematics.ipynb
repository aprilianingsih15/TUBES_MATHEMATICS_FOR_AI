```python
import pandas as pd
import numpy as np

# ===============================
# 1. LOAD DATASET (50 DATA)
# ===============================
data = pd.read_excel("/content/Dataset_Projek SVM.xlsx")

X = data.iloc[:50, 0:3].values
y = data.iloc[:50, 3].values  # label -1 / 1

# Normalisasi (samakan dengan Excel)
X = X / 100

# ===============================
# 2. PARAMETER (SAMA DENGAN EXCEL)
# ===============================
learning_rate = 0.01
C = 1
n_iter = 5

w = np.zeros(X.shape[1])
b = 0

# Untuk menyimpan hasil per iterasi
history = []

# ===============================
# 3. TRAINING MANUAL SVM
# ===============================
for it in range(1, n_iter + 1):
    total_loss = 0

    for i in range(len(X)):
        xi = X[i]
        yi = y[i]

        margin = yi * (np.dot(w, xi) + b)
        loss = max(0, 1 - margin)
        total_loss += loss

        if margin >= 1:
            grad_w = w
            grad_b = 0
        else:
            grad_w = w - C * yi * xi
            grad_b = -C * yi

        # Update
        w = w - learning_rate * grad_w
        b = b - learning_rate * grad_b

    avg_loss = total_loss / len(X)

    history.append([it, w[0], w[1], w[2], b, avg_loss])

    print(f"Iterasi {it}")
    print(f"w = {w}")
    print(f"b = {b}")
    print(f"Avg Hinge Loss = {avg_loss}")
    print("-" * 40)

# ===============================
# 4. HASIL AKHIR UNTUK PERBANDINGAN
# ===============================
result = pd.DataFrame(
    history,
    columns=["Iterasi", "w1", "w2", "w3", "b", "Avg_Hinge_Loss"]
)

print("\nHASIL AKHIR PER ITERASI:")
print(result)
```
